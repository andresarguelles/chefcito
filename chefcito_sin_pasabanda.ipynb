# Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
import seaborn as sns

from glob import glob
import librosa
import IPython.display as ipd

#Bibliotecas de configuracion de colores
from itertools import cycle

sns.set_theme(style="white", palette=None)
color_pal = plt.rcParams["axes.prop_cycle"].by_key()["color"]
color_cycle = cycle(plt.rcParams["axes.prop_cycle"].by_key()["color"])
```

# Leemos el audio

Nos limitamos a leer solo un segundo del audio

```python
audio_files = glob('dataset/*.wav')
y, sr = librosa.load(audio_files[0], sr=None)
y = y[:sr]
```

# Mostramos el audio

```python
plt.figure(figsize=(14, 5))
librosa.display.waveshow(y, sr=sr)
plt.title('Waveform del Audio')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.show()

#Reproducimos el audio que cargamos con librosa, el cual esta contenido en el array y y su sr es sr
ipd.Audio(data=y, rate=sr)
```

# Calculamos la Transformada rapida de Fourier

```python
y = y - np.mean(y) #Eliminamos el desplazamiento y centramos en cero las muestras de la seal
N = len(y)
fhat = np.fft.fft(y)  # Calculamos la FFT
PSD = np.abs(fhat)**2  # Calculamos la energa solo de la mitad positiva
 # Generamos el arreglo de frecuencias correspondiente (igual; la mitad) || freq = np.linspace(0, sr/2, N//2 + 1) 
freq = np.fft.fftfreq(N, d=1/sr)

# Plot
plt.figure(figsize=(12, 6))
plt.plot(freq[:N//2 + 1], PSD[:N//2 + 1])
plt.xlim(0, 120000)
plt.title('Espectro de Energa')
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Energa')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Buscamos cual es la frecuencia con ms energia

Aqui simplemente vemos cual de las frecuencias termina siendo ms relevante en cuanto a nivel de energia para la seal original.

```python
# Encontrar el ndice de la energa mxima
idx_max = np.argmax(PSD)

# Obtener la frecuencia correspondiente
freq_max = freq[idx_max]

# Y tambin la energa mxima si quieres
energia_max = PSD[idx_max]

print(f"La frecuencia con ms energa es {freq_max:.2f} Hz con una energa de {energia_max:.2f}")
```

# Eliminamos las frecuencias que no nos interesan para el analisis de las USVs

```python
fhat_filtrada = fhat.copy()
fhat_filtrada[np.abs(freq) < 20000] = 0
```

# Graficamos para corroborar que se eliminaron las frecuencias no deseadas

```python
N_filtrada = len(fhat_filtrada)

# Obtener energa espectral (magnitud al cuadrado)
energia_espectral = np.abs(fhat_filtrada[:N_filtrada//2 + 1])**2
frequencies = np.linspace(0, sr/2, N_filtrada//2 + 1)

plt.figure(figsize=(14, 6))
plt.plot(frequencies, energia_espectral)
plt.xlim(0, 120000)  # Puedes ajustar si quieres ms o menos rango
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Energa')
plt.title('Espectro de Frecuencias - Seal Filtrada (<20kHz Eliminadas)')
plt.grid(True)
plt.show()
```

# Comparacin grafica Original vs Filtrada

Mostramos una comparacin que nos permita ver que se ha realizado la limpieza correctamente.

```python
max_y = max(max(PSD), max(energia_espectral))

plt.figure(figsize=(18, 6))

# Seal cruda
plt.subplot(1, 2, 1)
plt.plot(freq, PSD)
plt.xlim(0, 120000)
plt.ylim(0, max_y)  # Usar la misma escala Y
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.title('Seal Cruda')
plt.grid(True)

# Seal filtrada
plt.subplot(1, 2, 2)
plt.plot(frequencies, energia_espectral, color='green')
plt.xlim(0, 120000)
plt.ylim(0, max_y)  # Misma escala Y
plt.xlabel('Frecuencia (Hz)')
plt.ylabel('Magnitud')
plt.title('Seal Filtrada')
plt.grid(True)

# Mostrar todo junto
plt.tight_layout()
plt.show()
```

# Corroboramos IFFT

Simplemente buscamos que el elemento complejo ms grande sea sper despreciable

```python
y_limpio = np.fft.ifft(fhat_filtrada)
print(np.max(np.abs(y_limpio.imag)))
```

# Reconstruimos la seal filtrada y la comparamos con la seal original

```python
# Asegrate de trabajar solo con la parte real
y_limpio_real = y_limpio.real

# Crear eje de tiempo
t = np.arange(len(y_limpio_real)) / sr

plt.figure(figsize=(14, 5))
librosa.display.waveshow(y, sr=sr, alpha=0.7, label='Original')
librosa.display.waveshow(y_limpio.real, sr=sr, color='r', alpha=0.7, label='Filtrado')
plt.title('Comparacin de Seal Original vs Filtrada')
plt.xlabel('Tiempo (s)')
plt.ylabel('Amplitud')
plt.legend()
plt.show()


# Dado que la seal fue filtrada para tener solo frecuencias mayores a 20kHz, esta seal no es posible de escuchar
ipd.Audio(data=y_limpio_real, rate=sr)
```
